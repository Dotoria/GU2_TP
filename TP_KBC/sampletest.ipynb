```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import Planck15 as cosmo
```

```python
data=np.loadtxt('GU2_TP_Test_BeomChan_Koh.csv',skiprows=1,delimiter=',').T
```

```python
z=data[1]
rmag=data[6]
```

```python
plt.figure(figsize=(10,8))
plt.scatter(z,rmag,c='black',s=0.1,alpha=0.3)
plt.xlabel("redshift",fontsize=14)
plt.ylabel("r band magnitude",fontsize=14)
plt.ylim(10,22)
plt.show()
```

```python
Mr= rmag - 5.0*np.log10(cosmo.luminosity_distance(z).to('pc').value/10.0)
```

```python
rcrit23=17.77 -5.0*np.log10(cosmo.luminosity_distance(0.23).value)-25
rcrit20=17.77 -5.0*np.log10(cosmo.luminosity_distance(0.2).value)-25
rcrit15=17.77 -5.0*np.log10(cosmo.luminosity_distance(0.15).value)-25
rcrit17=17.77 -5.0*np.log10(cosmo.luminosity_distance(0.17).value)-25
print(rcrit23)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(z,Mr,c='black',s=0.1,alpha=0.3)
plt.axhline(y=rcrit23,c='r')
plt.axhline(y=rcrit20,c='b')
plt.axhline(y=rcrit15,c='magenta')
plt.axhline(y=rcrit17,c='cyan')
plt.axvline(x=0.23,c='r')
plt.axvline(x=0.2,c='b')
plt.axvline(x=0.15,c='magenta')
plt.axvline(x=0.17,c='cyan')
plt.xlabel("redshift",fontsize=14)
plt.ylabel("Mr",fontsize=14)
plt.gca().invert_yaxis()
plt.xlim(0,0.3)
plt.ylim(-10,-25)
plt.show()
```
